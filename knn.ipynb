```python
df=pd.read_excel('/content/cancer patient data sets.xlsx')
```

```python
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
x=df.iloc[:,2:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
sc=StandardScaler()
sc.fit(x)
```

```python
x=sc.transform(x)
x
```

```python
# x=sc.fit_tranform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
y_train.shape
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x,y)
```

```python
y_predict=knn.predict(x_test)
```

```python
accuracy_score(y_test,y_predict)
```

```python
# predicting new value after fir transform is applied
new=sc.transform([[1,2,4,5,4,3,2,2,4,3,3,4,3,4,2,2,3,1,2,3,4,2]])
new
```

```python
knn.predict(new)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,y_predict))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

print(ConfusionMatrixDisplay.from_predictions(y_test,y_predict))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predict))
```
