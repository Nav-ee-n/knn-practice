```python
import pandas as pd
import numpy as np
```

```python
l=[[7,7,"Bad"],[7,4,"Bad"],[3,4,"Good"],[1,4,"Good"],[3,4,"?"]]
df=pd.DataFrame(l,columns=['x1','x2','y'])
df
```

```python
x=df.iloc[:,0:2]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=3)

knn.fit(x,y)
```

```python
knn.predict([[3,7]])
```
